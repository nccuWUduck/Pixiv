```python
頁數=700
like數=100
like_list = []
link_list = []
name_list = []

for i in range(1,頁數):
    print(i)
    if(i==1):
        url= 'https://www.pixiv.net/tags/ウマ娘プリティーダービー/novels'
    else:
        url= 'https://www.pixiv.net/tags/ウマ娘プリティーダービー/novels'+'?p='+str(i)
    #print(url)
    driver.get(url) # 更改網址以前往不同網頁
    driver.implicitly_wait(3)
    elements = driver.find_elements(By.CLASS_NAME, "sc-xy7s8e-1.gLICWZ")

# 获取元素的文本内容
    for element in elements:
        soup = BeautifulSoup(element.get_attribute("innerHTML"), 'html.parser')
        #print(soup)
        like_count = soup.find_all("span", {"class": "sc-eoqmwo-2 dfUmJJ"})
        #print(like_count)
        if(like_count):
            if(int(like_count[0].text.replace(',', ''))>like數):
                #invalid literal for int() with base 10: '1,351'
                #print(like_count[0].text)
                link = soup.find('a')['href']
                name_ele= soup.find_all("a", {"class": "sc-d98f2c-0 sc-iasfms-5 sc-1c4k3wn-14 eyLYWB ZgTBh gtm-novel-searchpage-result-title"})
                    
                print(like_count[0].text,'https://www.pixiv.net'+link)
                #https://www.pixiv.net/

                like_list.append(like_count[0].text)
                link_list.append('https://www.pixiv.net'+link)
                name_list.append(name_ele[0].text)       
    
```
